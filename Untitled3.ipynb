```python
clean_data=pd.read_csv("clean_dataframe.csv")
clean_data=clean_data.transpose()
clean_data.columns = clean_data.iloc[0]
clean_data=clean_data[1:]
clean_data=clean_data.transpose()
kmEngine = KMeans(n_clusters=4,init="k-means++", n_init=10, max_iter=300,tol=0.0001, random_state=200)
kmeans2=kmEngine.fit(clean_data)
clean_data["num"]= kmEngine.predict(clean_data)
sns.scatterplot(data=clean_data, x="Fish species, threatened", y="Bird species, threatened",hue="num")
clean_data[clean_data["num"]==0]
```

```python
sns.pairplot(clean_data, hue="num") 
```

```python
Z = sch.linkage(clean_data, method='complete', metric='euclidean')
fig = plt.figure(figsize=(20, 40))
dendro = sch.dendrogram(Z, orientation='left', leaf_rotation=0, leaf_font_size=15)
```

```python
fig = plt.figure(figsize=(10, 7))
plt.plot(Z[:,2],'o-')
plt.grid(axis='y') 
```

```python
nb_clust = 3
clusters = sch.fcluster(Z, nb_clust, criterion='maxclust')
crosstab=pd.crosstab(clean_data["num"],clusters)
proba = crosstab.values/crosstab.values.sum(axis=1, keepdims=True) # divide each element of a row by the sum of the row
print(proba)
entropy = [sp.stats.entropy(row, base=2) for row in proba]
print("entropy of each cluster: ", entropy)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
x=clean_data["Bird species, threatened"]
y=clean_data["Fish species, threatened"]
z=clean_data["Terrestrial and marine protected areas (% of total territorial area)"]
cmap = ListedColormap(sns.color_palette("husl", 256).as_hex())
ax.scatter(x,y,z,c=clean_data["num"])
plt.xlabel("bird")
plt.ylabel("fish")
plt.show()
```

```python
db_clust = DBSCAN(eps=1, min_samples=10).fit(clean_data)
p = sns.scatterplot(data=clean_data, x="Fish species, threatened", y="Bird species, threatened", hue=db_clust.labels_, legend="full", palette="deep")
sns.move_legend(p, "upper right", bbox_to_anchor=(1.17, 1.2), title='Clusters')
plt.show()
```

```python
# n_neighbors = 5 as kneighbors function returns distance of point to itself (i.e. first column will be zeros) 
nbrs = NearestNeighbors(n_neighbors=12).fit(clean_data)
# Find the k-neighbors of a point
neigh_dist, neigh_ind = nbrs.kneighbors(clean_data)
# sort the neighbor distances (lengths to points) in ascending order
# axis = 0 represents sort along first axis i.e. sort along row
sort_neigh_dist = np.sort(neigh_dist, axis=0)
k_dist = sort_neigh_dist[:, 4]
plt.plot(k_dist)
plt.ylabel("k-NN distance")
plt.xlabel("Sorted observations (4th NN)")
plt.show()
```
